```python
# Combine Home and Away DataFrames
cbb_game_df = pd.concat([home_df, away_df])
cbb_game_df
```

```python
# Read Stats Data for Season to Date
stats_df = pd.read_excel('cbb_training_data.xlsx')
stats_df.drop(columns=['Unnamed: 0'], inplace=True)

stats_df
```

```python
year = 2022
month = 11
# Get Current Season
if month < 10:
    year = year - 1
    stats_df = stats_df[stats_df['date'] >= f'{year}-06-01']
else:
    stats_df = stats_df[stats_df['date'] >= f'{year}-06-01']

stats_df
```

```python
stats_df.sort_values(by='date', ascending=True, inplace=True)
stats_df = stats_df.groupby(['team_id']).tail(1)
```

```python
# Merge stats with schedule df
game_data_df = pd.merge(cbb_game_df, stats_df, how='left', on='team_id')
game_data_df
```

```python
# Drop and Rename Columns
game_data_df.drop(columns=['game_id_y','date_y', 'team_y', 'opp_y', 'capacity_y', 'venue_id_y', 'opp_id_y', 'team_rank_y', 'opp_rank_y'], inplace=True)
game_data_df.rename(columns={'game_id_x': 'game_id','date_x': 'date', 'team_x':'team', 'opp_x': 'opp', 'capacity_x': 'capacity', 'venue_id_x': 'venue_id',
                              'opp_id_x': 'opp_id', 'team_rank_x': 'team_rank', 'opp_rank_x': 'opp_rank'}, inplace=True)

game_data_df
```

```python
# Normalize Ranking and Capacity Data
game_data_df['capacity'] = game_data_df['capacity']/game_data_df['capacity'].max()

game_data_df['team_rank'] = np.where(game_data_df['team_rank'] == 0, 50, game_data_df['team_rank'])
game_data_df['team_rank'] = game_data_df['team_rank'].fillna(50)

game_data_df['opp_rank'] = np.where(game_data_df['opp_rank'] == 0, 50, game_data_df['opp_rank'])
game_data_df['opp_rank'] = game_data_df['opp_rank'].fillna(50)

game_data_df['team_rank'] = game_data_df['team_rank']/game_data_df['team_rank'].max()
game_data_df['opp_rank'] = game_data_df['opp_rank']/game_data_df['opp_rank'].max()
#game_data_df.dropna(inplace=True)
game_data_df
```

# Predict Scores Based on Team Data

```python
model_path = '/Users/matthew.yough/Documents/GitHub/ncaab_game_predictor'
model_file = 'CBB_Score_Model.pkl'

regr = joblib.load(f"{model_path}/{model_file}")
```

```python
game_data_df_sub = game_data_df[['team', 'home_game', 'away_game','opp', 'capacity', 'team_rank', 'opp_rank', 'total_team_points','total_opp_points', 'total_team_fg_att',
                   'total_opp_fg_att', 'total_team_fg_made', 'total_opp_fg_made', 'total_team_fg_pct', 'total_opp_fg_pct',
                   'total_team_3pt_att','total_opp_3pt_att', 'total_team_3pt_made', 'total_opp_3pt_made', 'total_team_3pt_pct', 'total_opp_3pt_pct',
                    'total_team_ft_att','total_opp_ft_att', 'total_team_ft_made', 'total_opp_ft_made', 'total_team_ft_pct', 'total_opp_ft_pct', 
                    'total_team_rebounds', 'total_opp_rebounds', 'total_team_assists', 'total_opp_assists', 'total_team_steals', 'total_opp_steals',
                    'total_team_blocks', 'total_opp_blocks', 'total_team_turnovers', 'total_opp_turnovers', 'total_team_fouls', 'total_opp_fouls',
                    'venue_code', 'team_code', 'opp_code']]
game_data_df_sub.dropna(inplace=True)

X = game_data_df_sub[['capacity', 'team_rank', 'opp_rank', 'total_team_points','total_opp_points', 'total_team_fg_att',
                   'total_opp_fg_att', 'total_team_fg_made', 'total_opp_fg_made', 'total_team_fg_pct', 'total_opp_fg_pct',
                   'total_team_3pt_att','total_opp_3pt_att', 'total_team_3pt_made', 'total_opp_3pt_made', 'total_team_3pt_pct', 'total_opp_3pt_pct',
                    'total_team_ft_att','total_opp_ft_att', 'total_team_ft_made', 'total_opp_ft_made', 'total_team_ft_pct', 'total_opp_ft_pct', 
                    'total_team_rebounds', 'total_opp_rebounds', 'total_team_assists', 'total_opp_assists', 'total_team_steals', 'total_opp_steals',
                    'total_team_blocks', 'total_opp_blocks', 'total_team_turnovers', 'total_opp_turnovers', 'total_team_fouls', 'total_opp_fouls',
                    'venue_code', 'team_code', 'opp_code']]
X = X.reset_index(drop=True)
```

```python
# Run Model
y_pred = regr.predict(X)
```
