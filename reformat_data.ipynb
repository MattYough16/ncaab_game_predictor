```python
import pandas as pd
```

```python
df = pd.read_excel('cbb_norm_data.xlsx')
df.drop(columns=['Unnamed: 0'], inplace=True)
df
```

```python

# convert the 'Date' column to datetime format
df['date']= pd.to_datetime(df['date'])
```

```python
# Summarize Data Across a Season
df.sort_values(by=['season','date'], ascending=True, inplace=True)
df
```

```python
df['game_counter'] = 1
df['team_pace'] = df.groupby(['season','team'])['pace'].cumsum()/df.groupby(['season','team'])['game_counter'].cumsum() * 100
df['opp_pace'] = df.groupby(['season','opp'])['pace'].cumsum()/df.groupby(['season','opp'])['game_counter'].cumsum() * 100
df.to_excel('pre_totaled_data.xlsx')

df['total_team_games'] = df.groupby(['season','team'])['games'].cumsum()
df['total_opp_games'] = df.groupby(['season','opp'])['games'].cumsum()
df['total_team_points'] = df.groupby(['season','team'])['team_points'].cumsum()
df['total_opp_points'] = df.groupby(['season','opp'])['opp_points'].cumsum()
df['total_team_fg_att'] = df.groupby(['season','team'])['team_field_goal_att'].cumsum()
df['total_opp_fg_att'] = df.groupby(['season','opp'])['opp_field_goal_att'].cumsum()
df['total_team_fg_made'] = df.groupby(['season','team'])['team_field_goal_made'].cumsum()
df['total_opp_fg_made'] = df.groupby(['season','opp'])['opp_field_goal_made'].cumsum()
df['total_team_fg_pct'] = df.groupby(['season','team'])['team_field_goal_pct'].cumsum()/df.groupby(['season','team'])['game_counter'].cumsum() * 100
df['total_opp_fg_pct'] = df.groupby(['season','opp'])['opp_field_goal_pct'].cumsum()/df.groupby(['season','opp'])['game_counter'].cumsum() * 100
df['total_team_3pt_att'] = df.groupby(['season','team'])['team_3pt_att'].cumsum()
df['total_opp_3pt_att'] = df.groupby(['season','opp'])['opp_3pt_att'].cumsum()
df['total_team_3pt_made'] = df.groupby(['season','team'])['team_3pt_made'].cumsum()
df['total_opp_3pt_made'] = df.groupby(['season','opp'])['opp_3pt_made'].cumsum()
df['total_team_3pt_pct'] = df.groupby(['season','team'])['team_3pt_pct'].cumsum()/df.groupby(['season','team'])['game_counter'].cumsum() * 100
df['total_opp_3pt_pct'] = df.groupby(['season','opp'])['opp_3pt_pct'].cumsum()/df.groupby(['season','opp'])['game_counter'].cumsum() * 100
df['total_team_ft_att'] = df.groupby(['season','team'])['team_free_throw_att'].cumsum()
df['total_opp_ft_att'] = df.groupby(['season','opp'])['opp_free_throw_att'].cumsum()
df['total_team_ft_made'] = df.groupby(['season','team'])['team_free_throw_made'].cumsum()
df['total_opp_ft_made'] = df.groupby(['season','opp'])['opp_free_throw_made'].cumsum()
df['total_team_ft_pct'] = df.groupby(['season','team'])['team_free_throw_pct'].cumsum()/df.groupby(['season','team'])['game_counter'].cumsum() * 100
df['total_opp_ft_pct'] = df.groupby(['season','opp'])['opp_free_throw_pct'].cumsum()/df.groupby(['season','opp'])['game_counter'].cumsum() * 100
df['total_team_rebounds'] = df.groupby(['season','team'])['team_rebounds'].cumsum()
df['total_opp_rebounds'] = df.groupby(['season','opp'])['opp_rebounds'].cumsum()
df['total_team_assists'] = df.groupby(['season','team'])['team_assists'].cumsum()
df['total_opp_assists'] = df.groupby(['season','opp'])['opp_assists'].cumsum()
df['total_team_steals'] = df.groupby(['season','team'])['team_steals'].cumsum()
df['total_opp_steals'] = df.groupby(['season','opp'])['opp_steals'].cumsum()
df['total_team_blocks'] = df.groupby(['season','team'])['team_blocks'].cumsum()
df['total_opp_blocks'] = df.groupby(['season','opp'])['opp_blocks'].cumsum()
df['total_team_turnovers'] = df.groupby(['season','team'])['team_turnovers'].cumsum()
df['total_opp_turnovers'] = df.groupby(['season','opp'])['opp_turnovers'].cumsum()
df['total_team_fouls'] = df.groupby(['season','team'])['team_fouls'].cumsum()
df['total_opp_fouls'] = df.groupby(['season','opp'])['opp_fouls'].cumsum()

df
```

```python
df['team_ppg'] = df['total_team_points']/df['total_team_games']
df['opp_ppg'] = df['total_opp_points']/df['total_opp_games']
df['team_fg_att_pg'] = df['total_team_fg_att']/df['total_team_games']
df['opp_fg_att_pg'] = df['total_opp_fg_att']/df['total_opp_games']
df['team_fg_made_pg'] = df['total_team_fg_made']/df['total_team_games']
df['opp_fg_made_pg'] = df['total_opp_fg_made']/df['total_opp_games']
df['team_3pt_att_pg'] = df['total_team_3pt_att']/df['total_team_games']
df['opp_3pt_att_pg'] = df['total_opp_3pt_att']/df['total_opp_games']
df['team_3pt_made_pg'] = df['total_team_3pt_made']/df['total_team_games']
df['opp_3pt_made_pg'] = df['total_opp_3pt_made']/df['total_opp_games']
df['team_ft_att_pg'] = df['total_team_ft_att']/df['total_team_games']
df['opp_ft_att_pg'] = df['total_opp_ft_att']/df['total_opp_games']
df['team_ft_made_pg'] = df['total_team_ft_made']/df['total_team_games']
df['opp_ft_made_pg'] = df['total_opp_ft_made']/df['total_opp_games']
df['team_rebounds_pg'] = df['total_team_rebounds']/df['total_team_games']
df['opp_rebounds_pg'] = df['total_opp_rebounds']/df['total_opp_games']
df['team_assists_pg'] = df['total_team_assists']/df['total_team_games']
df['opp_assists_pg'] = df['total_opp_assists']/df['total_opp_games']
df['team_steals_pg'] = df['total_team_steals']/df['total_team_games']
df['opp_steals_pg'] = df['total_opp_steals']/df['total_opp_games']
df['team_blocks_pg'] = df['total_team_blocks']/df['total_team_games']
df['opp_blocks_pg'] = df['total_opp_blocks']/df['total_opp_games']
df['team_turnovers_pg'] = df['total_team_turnovers']/df['total_team_games']
df['opp_turnovers_pg'] = df['total_opp_turnovers']/df['total_opp_games']
df['team_fouls_pg'] = df['total_team_fouls']/df['total_team_games']
df['opp_fouls_pg'] = df['total_opp_fouls']/df['total_opp_games']
```

```python
df.to_excel('cbb_norm_data.xlsx')
```
